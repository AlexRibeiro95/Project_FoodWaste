```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\USER\\Documents\\GitHub\\Project1\\FAOSTAT_data_en_7-8-2024 (1).csv')
```

```python
df
```

```python
df.columns = df.columns.str.lower()
```

```python
#replace the spaces for underscores
df.columns = df.columns.str.replace(' ', '_')
```

```python
df_filtered = df
```

```python
#change the name of the columns area to country
df_filtered = df_filtered.rename(columns = {'area': 'country'})
```

```python
df_filtered
```

```python
df_filtered = df_filtered.drop(['domain', 'domain_code', 'element_code', 'item_code', 'year_code', 'source_code', 'source', 'flag', 'note', 'flag_description'], axis=1)
```

```python
df_filtered["value"] = df_filtered["value"].round(2)
```

```python
df_filtered
```

```python
df_filtered = df_filtered[df_filtered['country'].isin(list_countries)]
```

```python
df_filtered.reset_index(drop=True, inplace=True)
```

```python
df_filtered
```

```python
df_filtered["element"].value_counts()
```

```python
element_mapping = {
    'Direct emissions (N2O)': 'Nitrous Oxide (N2O)',
    'Emissions (N2O)': 'Nitrous Oxide (N2O)',
    'Emissions (CO2eq) from N2O (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2eq) (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CH4)': 'Methane (CH4)',
    'Emissions (CO2eq) from CH4 (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2eq) from F-gases (AR5)': 'Carbon Dioxide (CO2)'
}
df_filtered["element"] = df_filtered["element"].replace(element_mapping)
```

```python
df_filtered
```

```python
#Condense the data frame to show just 1 entry per country and year with the average of value grouped by element and item
df_filtered = df_filtered.groupby(['country', 'year', 'element', 'item', 'unit']).agg({'value':'mean'}).reset_index().round(2)  
```
